```python
from tkinter import *
main = Tk()
number = StringVar()
numberTemp = []
numberTemp.append('0')
number1 = 0
number2 = 0
activeOperator = None

def appendFinalArray(input):
    global numberTemp
    #Default 0 handling
    if(input == 0 and numberTemp[-1] == 0):
        return None

    #Decimal place 0 Handling
    if(input == ".0" and numberTemp[-1] == '0'):
        input = "0.0"
    if(numberTemp[-1] == ".0"):
        numberTemp[-1] = '.'
    if(numberTemp[-1] == "0.0"):
        numberTemp[-1] = "0."

    #Clearing base 0
    if(len(numberTemp) == 1 and numberTemp[-1] == '0'):
        numberTemp = []
    numberTemp.append(input)

# C BUTTON
def clearFinalArray():
    global numberTemp
    numberTemp.clear()
    numberTemp.append('0')
    update()

"""

"""

def operator(input):
    global number2,number1, activeOperator, numberTemp
    number1 = float(number.get())
    numberTemp = ['0']

    activeOperator = input

def calculate():
    global number2, number1, number, numberTemp
    number2 = float(number.get())
    match (activeOperator):
        case '+':
            result = number1 + number2
        case '-':
            result = number1 - number2
        case '*':
            result = number1 * number2
        case '/':
            result = number1 / number2
        case '^':
            result = number1 ** number2
    clearFinalArray()
    appendFinalArray(str(result))
    # temp = number1
    # number1 = number2
    # number2 = temp
    
    number.set(str(result))
    

Button(main, text='0', height=5, width = 12, command = lambda: appendFinalArray('0')).grid(row=5, column = 0, columnspan=2)

Button(main, text='1', height=5, width = 5, command = lambda: appendFinalArray('1')).grid(row=4, column = 0)
Button(main, text='2', height=5, width = 5, command = lambda: appendFinalArray('2')).grid(row=4, column = 1)
Button(main, text='3', height=5, width = 5, command = lambda: appendFinalArray('3')).grid(row=4, column = 2)

Button(main, text='4', height=5, width = 5, command = lambda: appendFinalArray('4')).grid(row=3, column = 0)
Button(main, text='5', height=5, width = 5, command = lambda: appendFinalArray('5')).grid(row=3, column = 1)
Button(main, text='6', height=5, width = 5, command = lambda: appendFinalArray('6')).grid(row=3, column = 2)

Button(main, text='7', height=5, width = 5, command = lambda: appendFinalArray('7')).grid(row=2, column = 0)
Button(main, text='8', height=5, width = 5, command = lambda: appendFinalArray('8')).grid(row=2, column = 1)
Button(main, text='9', height=5, width = 5, command = lambda: appendFinalArray('9')).grid(row=2, column = 2)

Button(main, text='=', height=5, width = 5, command = calculate).grid(row=5, column = 3)
Button(main, text='+', height=5, width = 5, command = lambda: operator('+')).grid(row=4, column = 3)
Button(main, text='-', height=5, width = 5, command = lambda: operator('-')).grid(row=3, column = 3)
Button(main, text='/', height=5, width = 5, command = lambda: operator('/')).grid(row=2, column = 3)
Button(main, text='*', height=5, width = 5, command = lambda: operator('*')).grid(row=1, column = 3)

Button(main, text='+/-', height=5, width = 5).grid(row=1, column = 0)
Button(main, text='^', height=5, width = 5, command = lambda: operator('^')).grid(row=1, column = 1)
Button(main, text='C', height=5, width = 5, command = clearFinalArray).grid(row=1, column = 2)


Button(main, text='.', height=5, width = 5, command = lambda: appendFinalArray(".0")).grid(row=5, column = 2)

def update():
    number.set("".join(numberTemp))
    main.after(50,update)
main.after(50,update)
answerField = Label(main, textvariable=number).grid(row=0,column = 0, columnspan=3)
main.mainloop()
```
